```python
from playwright.async_api import async_playwright
import asyncio

async def scrape_all_endangered_species():
    data = []
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto("https://awionline.org/content/list-endangered-species")
        tabs = [
            "#mammals", "#birds", "#amphibians", "#reptiles", 
            "#fishes", "#arthropods", "#mollusks", "#invertebrates", "#plants"
        ]

        for tab in tabs:
            await page.goto(f"https://awionline.org/content/list-endangered-species{tab}")
            await page.wait_for_selector("table tr")
            group_name = tab.replace("#", "").capitalize()

            rows = await page.query_selector_all("table tr")
            for row in rows:
                cells = await row.query_selector_all("td")
                text = [await cell.inner_text() for cell in cells]
                if text:
                    if len(text) == 3:
                        text.append(group_name)
                    data.append(text)

        await browser.close()

    df = pd.DataFrame(data, columns=["Common Name", "Scientific Name", "Status", "Group"])
    df.to_csv("all_endangered_species.csv", index=False)
    print("all_endangered_species.csv")

await scrape_all_endangered_species()
```
